```python
import pandas as pd

url = "https://raw.githubusercontent.com/maxmilhanaguiar/NF3-BIG-DATA-Maxmilhan/main/cursos-prouni.csv"

df = pd.read_csv(url)

# LETRA A #
print ("=================================== LETRA A ===================================\n")
df.fillna(0.0, inplace=True)
print(df)

# LETRA B #
print ("=================================== LETRA B ===================================\n")
grupo_grau = df.groupby('grau')
print(grupo_grau)

# LETRA C #
print ("=================================== LETRA C ===================================\n")
cursos_interesse = ['Matemática', 'Medicina', 'Pedagogia']
cursos = df[df['curso_busca'].isin(cursos_interesse)].groupby('curso_busca')
print(cursos)

# LETRA D #
print ("=================================== LETRA D ===================================\n")
grupo_estado = df.groupby('uf_busca')
m_n_estado = grupo_estado['nota_integral_cotas'].mean()
print(m_n_estado)

# LETRA E #
print ("=================================== LETRA E ===================================\n")
tecnologicos = df[df['grau'] == 'Tecnológico'].groupby('curso_busca')
print(tecnologicos)

# LETRA F #
print ("=================================== LETRA F ===================================\n")
df.drop('cidade_filtro', axis=1, inplace=True)
print(df)

# LETRA G #
print ("=================================== LETRA G ===================================\n")
mensalidades_medicina = df[df['curso_busca'] == 'Medicina']['mensalidade'].mean()
print(mensalidades_medicina)

# LETRA H #
print ("=================================== LETRA H ===================================\n")
m_n_integral = df[df['turno'] == 'Integral']['nota_integral_cotas'].mean()
print(m_n_integral)

# LETRA I #
print ("=================================== LETRA I ===================================\n")
estat_bacharelado = df[df['grau'] == 'Bacharelado']['nota_integral_ampla'].describe()
print(estat_bacharelado)

# LETRA J #
print ("=================================== LETRA J ===================================\n")
import matplotlib.pyplot as plt

grupo_grau['nota_integral_cotas'].mean().plot(kind='bar')
plt.xlabel('Grau')
plt.ylabel('Média das Notas Integral de Cotas')
plt.title('Comparação das Notas Integral de Cotas por Grau dos Cursos')
plt.show()
```
